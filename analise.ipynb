```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura dos dados
df = pd.read_excel('dados_tratados8.0.xlsx')

# Colunas de interesse
colunas_faixa_etaria = [
    "Número aproximado de crianças:",
    "Número aproximado de jovens:",
    "Número aproximado de adultos:",
    "Número aproximado de idosos:"
]

# Soma total de cada faixa etária
totais_faixa_etaria = df[colunas_faixa_etaria].sum()

# Cores personalizadas
cores = ['skyblue', 'orange', 'mediumseagreen', 'orchid']

# Plotagem
plt.figure(figsize=(8, 5))
totais_faixa_etaria.plot(kind="bar", color=cores)
plt.title("Total de Participantes por Faixa Etária")
plt.xlabel("Faixa Etária")
plt.ylabel("Quantidade Total")
plt.xticks(rotation=45)
plt.grid(axis="y", color="black")
plt.tight_layout()
plt.show()
```

```python
zoneamento_participantes = df.groupby("Zoneamento")["Número aproximado de participantes:"].sum()

zoneamento_participantes.plot(
    kind="pie", autopct='%1.1f%%', startangle=90,
    colors=["mediumseagreen", "gold"], figsize=(6, 6), title="Distribuição de Participantes por Zoneamento"
)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
cidade_participantes = df.groupby("Cidade")["Número aproximado de participantes:"].sum().sort_values()

cidade_participantes.plot(kind="barh", figsize=(8, 10), color="steelblue", title="Participantes por Cidade")
plt.xlabel("Número de Participantes")
plt.tight_layout()
plt.show()
```

```python
zoneamento_lixo = df.groupby("Zoneamento")["Número total de sacos de lixo (cheios):"].sum()

zoneamento_lixo.plot(kind="bar", color=["mediumseagreen", "gold"], figsize=(6, 5))
plt.title("Total de Sacos de Lixo por Zoneamento")
plt.ylabel("Sacos de Lixo")
plt.xlabel("Zoneamento")
plt.grid(axis="y", color="black")
plt.tight_layout()
plt.show()
```

```python
cidade_lixo = df.groupby("Cidade")["Número total de sacos de lixo (cheios):"].sum().sort_values(ascending=False).head(10)

cidade_lixo.plot(kind="bar", color="darkorange", figsize=(10, 5), title="Top 10 Cidades - Sacos de Lixo Coletados")
plt.ylabel("Sacos de Lixo")
plt.xticks(rotation=45)
plt.grid(axis="y", color="black")
plt.tight_layout()
plt.show()
```

```python
estado_peso = df.groupby("Estado")["Peso total dos resíduos (kg)"].sum().sort_values()

estado_peso.plot(kind="barh", figsize=(8, 6), color="firebrick", title="Peso Total de Resíduos por Estado (kg)")
plt.xlabel("Peso (kg)")
plt.tight_layout()
plt.show()
```
